```python
from RungeKutteM import RungeKutteMethod
from AdamsM import adams_method
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from runge_rule import runge_rule
from CaushiM import eilerMethod
```

```python
delta = 10
r = 1
b = 8/3
```

```python

def xfunc(t, x, y, z):
    return delta * (y - x)    
def yfunc(t, x, y, z):
    return x * (r - z) - y
def zfunc(t, x, y, z):
    return x * y - b * z
```

```python
def solving_with_graph(start_values: list, amount_of_points: int):
    k = RungeKutteMethod([xfunc, yfunc, zfunc], 0, 1, start_values, amount_of_points)
    a = np.linspace(0, 1, amount_of_points)
    my = []
    my1 = []
    my2 = []
    for i in range(len(k)): 
        my.append(k[i][1])
        my1.append(k[i][0])
        my2.append(k[i][2])

    plt.plot(a, my, color = 'orange')
    plt.plot(a, my1, color = 'green')
    plt.plot(a, my2, color = 'purple')
    plt.scatter(a,my, color = "green")
    plt.scatter(a, my1, color = 'black')
    plt.show()
    plt.close()

    fig = plt.figure()
    ax = fig.add_subplot(projection='3d')

        # !!!
    ax.plot3D(my, my1, my2, 'gray')

    plt.show()

    k = eilerMethod([xfunc, yfunc, zfunc], 0, 1, start_values, amount_of_points)

    my = []
    my1 = []
    my2 = []
    for i in range(len(k)): 
        my.append(k[i][1])
        my1.append(k[i][0])
        my2.append(k[i][2])

    plt.plot(a, my, color = 'orange')
    plt.plot(a, my1, color = 'green')
    plt.plot(a, my2, color = 'purple')
    plt.scatter(a,my, color = "green")
    plt.scatter(a, my1, color = 'black')
    plt.show()
    plt.close()



    k = adams_method([xfunc, yfunc, zfunc], 0, 1, start_values, amount_of_points)

    my = []
    my1 = []
    my2 = []
    for i in range(len(k)): 
        my.append(k[i][1])
        my1.append(k[i][0])
        my2.append(k[i][2])

    plt.plot(a, my, color = 'orange')
    plt.plot(a, my1, color = 'green')
    plt.plot(a, my2, color = 'purple')
    plt.scatter(a,my, color = "green")
    plt.scatter(a, my1, color = 'black')
    plt.show()
    plt.close()
```

Тут видно, что при начальных значениях 0,0,0 решение системы будет тождестевнный 0. Потому что ни 1 из методов не может пройти это значение в силу того как они написаны. Поэтому далее будем использовать значения, приближенные к 0

```python
solving_with_graph([0, 0, 0], 1280)

```

Тут видно, что метод Рунге Кутты для небольшого количества точек работает лучше, чем остальные. Поэтому все дальнейшие тесты я буду проводить на нём. 

```python
solving_with_graph([0.1, 0.1, 0.1], 128)
```

Также, я решил 1 из начальных значений поменять на 10, чтобы графики не сливались в процессе. И исследовать поведение параметров для этих значений

```python
solving_with_graph([0.1, 10, 0.1], 128)
```
